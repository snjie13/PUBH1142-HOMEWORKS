### Question One

 A spreadsheet is mainly used for organizing data in rows and columns and doing calculations on smaller datasets. A database is built to store and manage large amounts of structured data, often across multiple related tables, with better security and integrity controls.

### Question Two

 Spreadsheets are easy to use, good for small datasets, and useful for quick calculations and charts. However, they are not ideal for large datasets, multiple users, or situations where strong data security and accuracy controls are needed.

### Question Three

 Databases can store large amounts of information, maintain relationships between data, allow multiple users, and provide stronger security. On the downside, they require more technical knowledge and setup compared to spreadsheets."
